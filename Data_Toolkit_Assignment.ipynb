Answer 01
- NumPy (Numerical Python) is a foundational library for numerical computing in
  Python.
- It provides support for large, multi-dimensional arrays and matrices, along
  with a collection of mathematical functions to operate on these arrays.

```python
# Example

import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Perform element-wise operations
arr_squared = arr ** 2
arr_sum = np.sum(arr)

print("Original Array:", arr)
print("Squared Array:", arr_squared)
print("Sum of Array:", arr_sum)
```

Answer 02
- Broadcasting in NumPy refers to the ability to perform arithmetic operations  
  on arrays of different shapes. NumPy automatically expands the smaller array to match the shape of the larger one, following specific rules:
  Python Tutorial.
- If the arrays have different dimensions, the shape of the smaller-dimensional
  array is padded with ones on the left side until both shapes have the same length.
- Two dimensions are compatible when:
  - They are equal, or
  - One of them is 1.
- This mechanism allows for efficient element-wise operations without the need for explicit loops.

```python
# Example

# !pip install numpy
import numpy as np

array_2d = np.array([[1, 2, 3], [4, 5, 6]])  # 2D array
scalar = 10  # Scalar value

result = array_2d + scalar
print(result)
```

Answer 04
- The groupby() method in Pandas is used to split data into groups based on  
  some criteria, apply a function to each group independently, and then combine the results back into a DataFrame. This is particularly useful for aggregation, transformation, and filtering operations.

```python
# Example

import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)
grouped = df.groupby('Category').sum()
print(grouped)
```

Answer 06
- NumPy Arrays
 - Homogeneity: All elements must be of the same type.
 - Performance: Optimized for numerical operations.
 - Memory Efficiency: More memory-efficient.
 - Functionality: Supports vectorized operations and broadcasting.
- Python Lists
 - Homogeneity: Elements can be of different types.
 - Performance: Slower for numerical operations.
 - Memory Efficiency: MempLess memory-efficient.
 - Functionality: Limited support for vectorized operations.

```python
# Example

import numpy as np

# Python List
py_list = [1, 2, 3, 4, 5]

# NumPy Array
np_array = np.array([1, 2, 3, 4, 5])

# Element-wise operation
py_list_squared = [x**2 for x in py_list]
np_array_squared = np_array**2

print("Python List Squared:", py_list_squared)
print("NumPy Array Squared:", np_array_squared)
```

Answer 13
- Handling missing data is crucial because it ensures the integrity and  
  accuracy of your analyses. Missing values can lead to biased results, errors in computations, and incorrect conclusions.

```python
# Example

import pandas as pd
import numpy as np

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, np.nan, 30, 35],
        'Salary': [50000, 60000, np.nan, 70000]}

df = pd.DataFrame(data)
print(df.fillna({'Age': df['Age'].mean(), 'Salary': df['Salary'].mean()}))
```

Answer 17
- Both apply() and map() are used to apply functions to data, but they differ  
  in their usage:
 - apply(): Used on DataFrames and Series. It applies a function along the axis
   of the DataFrame (rows or columns).
 - map(): Used only on Series. It maps each value in the Series to a
   corresponding value using a dictionary, Series, or function.

```python
# Example

import pandas as pd

# Sample Series
s = pd.Series(['cat', 'dog', 'rabbit'])

# Mapping using a dictionary
s_mapped = s.map({'cat': 'kitten', 'dog': 'puppy'})
print(s_mapped)
```

Answer 19
- Pandas provides robust tools for time series analysis:
 - Datetime Indexing: Allows for easy selection and manipulation of data based
   on time.
 - Resampling: Facilitates changing the frequency of time series data (e.g.,
   from daily to monthly).
 - Rolling Windows: Enables the computation of statistics over a moving window.
 - Time Zone Handling: Supports conversion between time zones.

```python
# Example

import pandas as pd

# Creating a time series
dates = pd.date_range('20230101', periods=6)
df = pd.DataFrame({'value': [1, 2, 3, 4, 5, 6]}, index=dates)

# Resampling to monthly frequency
monthly_df = df.resample('M').sum()
print(monthly_df)
```

Answer 20
- A pivot table in Pandas is used to reshape data, providing a way to summarize
 and aggregate data:
  - Aggregation: Computes summary statistics (e.g., mean, sum) for different
  subsets of data.
  - Reshaping: Transforms long-format data into wide-format data.

```python
# Example

import pandas as pd

# Sample data
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Category': ['A', 'B', 'A', 'B'],
    'Sales': [100, 200, 150, 250]
}

df = pd.DataFrame(data)

# Creating a pivot table
pivot_df = pd.pivot_table(df, values='Sales', index='Date', columns='Category', aggfunc='sum')
print(pivot_df)
```

Answer 21
- NumPy arrays are implemented in C and have a fixed size and type, allowing  
  for more efficient memory access and operations. In contrast, Python lists are dynamic and heterogeneous, leading to overhead in memory management and slower performance.

```python
# Example

import numpy as np
import time

# NumPy array slicing
arr = np.arange(1000000)
start_time = time.time()
arr_slice = arr[500:1000]
print("NumPy slicing time:", time.time() - start_time)

# Python list slicing
lst = list(range(1000000))
start_time = time.time()
lst_slice = lst[500:1000]
print("Python list slicing time:", time.time() - start_time)
```
